```python
import numpy as np
import seaborn as sns
import pandas as pd


import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv(r'C:\Users\Clint\Downloads\movies.csv.zip')
```

```python
df.head()
```

```python
df  = df.dropna()
```

```python
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df.head()
```

```python
pd.set_option('display.max_rows', None)
df.head()
```

```python
plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Correlation')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for movies')

plt.show()
```

```python
df  = df.sort_values(by=['gross'], inplace = False, ascending = False)
df.head()
```

```python
sns.regplot(x='budget',y='gross', data=df, scatter_kws={"color":"black"} , line_kws = {"color":"yellow"})
```

```python
df.corr(numeric_only=True , method = 'pearson')
```

```python
correlation_matrix = df.corr(numeric_only=True , method = 'pearson')

sns.heatmap(correlation_matrix,annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True , method = 'pearson')

sns.heatmap(correlation_matrix,annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_matrix = df_numerized.corr()
corr_pairs = correlation_matrix.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
